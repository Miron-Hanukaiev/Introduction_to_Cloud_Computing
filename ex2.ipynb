```python
# @title class ex2
import json
import ipywidgets as widgets
from IPython.display import display, HTML

# Load JSON file
json_path = "/content/drive/MyDrive/students.json"

with open(json_path, "r") as file:
    students = json.load(file)

display(HTML("""
  <style>
    body {
      background-color: #3b2a77;
      color: #f0eaff;
      font-family: 'Segoe UI', sans-serif;
    }
    h3 {
      color: #d1c4e9;
    }
    p {
      color: #dcd0ff;
    }
    .widget-label {
      color: #e1bee7 !important;
      font-weight: bold;
    }
    input, textarea {
      background-color: #5e4aa1 !important;
      color: #ffffff !important;
      border: 1px solid #b39ddb !important;
    }
    .widget-dropdown > select {
      background-color: #5e4aa1 !important;
      color: #ffffff !important;
      border: 1px solid #b39ddb !important;
    }
  </style>

  <h3>Students Data</h3>
  <p>Choose a student, and the details will appear below:</p>
"""))

# Dropdown options
student_names = [f"{s['name']} {s['last_name']}" for s in students]

dropdown = widgets.Dropdown(
    options=student_names,
    value=None,
    description='Student:',
    style={'description_width': 'initial'},
    layout=widgets.Layout(width='400px')
)

email_box = widgets.Textarea(
    value='',
    description='Email:',
    layout=widgets.Layout(width='600px', height='40px')
)

courses_box = widgets.Textarea(
    value='',
    description='Courses:',
    layout=widgets.Layout(width='600px', height='120px')
)

link_box = widgets.Textarea(
    value='',
    description='Link:',
    layout=widgets.Layout(width='600px', height='40px')
)

fav_course_box = widgets.Text(
    value='',
    placeholder='Type your favorite course...',
    description='Favorite:',
    layout=widgets.Layout(width='600px')
)

update_button = widgets.Button(
    description='Update JSON',
    button_style='success',
    tooltip='Save favorite course to JSON',
    layout=widgets.Layout(width='200px')
)

# When student selected → fill fields
def update_fields(change):
    full_name = change["new"]
    first, last = full_name.split()

    for s in students:
        if s["name"] == first and s["last_name"] == last:
            email_box.value = s["email"]
            courses_box.value = "\n".join(s["courses"])
            link_box.value = s["link"]
            fav_course_box.value = s.get("favorite", "")

dropdown.observe(update_fields, names='value')

display(dropdown, email_box, courses_box, link_box)

# Save update to JSON
def update_json(b):
    full_name = dropdown.value
    favorite = fav_course_box.value.strip()

    if not full_name:
        print("Please select a student first.")
        return
    if not favorite:
        print("Please type a favorite course first.")
        return

    first, last = full_name.split()
    for s in students:
        if s["name"] == first and s["last_name"] == last:
            s["favorite"] = favorite

    with open(json_path, "w") as file:
        json.dump(students, file, indent=4)

    print(f"{full_name}'s favorite '{favorite}' was saved successfully.")

update_button.on_click(update_json)

display(fav_course_box, update_button)
```
